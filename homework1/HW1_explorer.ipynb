## Problem 1

`sold_shares[i]` is the number of shares sold of stock `i + 1`. The solver minimizes the expected one-year value sold while raising exactly $30,000 after taxes and transaction costs.

```python
stock_solution = solve_stock_sales()

for stock_number, sold in enumerate(stock_solution.sold_shares, start=1):
    remaining = stock_solution.remaining_shares[stock_number - 1]
    print(f"Stock {stock_number}: sell {sold:.6f}, keep {remaining:.6f}")

print(f"Net cash raised: {stock_solution.net_cash:.2f}")
print(f"Expected one-year remaining value: {stock_solution.expected_remaining_value:.2f}")
```
